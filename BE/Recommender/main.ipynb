```python
import pandas as pd
import sklearn
```

- Ưu tiên khu vục "region"
- Tìm các post gần nhất bằng Euclidean Distance

```python

df = pd.read_csv("data.csv")
df
```

Step 1: 
    - Tách dataset, ưu tiên khu vực "region"
    - Tạo Minmax Scaler

```python
cal_cols = ["area_by_m2", "width_of_facade", "width_of_road", "is_legal", "price"]
user_last_visited = pd.DataFrame({
    "id": [50],
    "region": ["Lộc Tiến"],
    "area_by_m2": [100],
    "width_of_facade": [10],
    "width_of_road": [5],
    "is_legal": [1],
    "price": [5]
})
df_scaled_data = df.copy()
# df = df.append(user_last_visited.iloc[0])
df = pd.concat([df, user_last_visited])
user_last_visited
df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(df[cal_cols])

df_scaled_data = df.copy()

df_scaled_data[cal_cols] = scaler.transform(df[cal_cols])
user_last_visited[cal_cols] = scaler.transform(user_last_visited[cal_cols])

df_scaled_data
```

Step 2: 
    - Lọc ra region từ lần cuối truy cập của user
    - Tính Euclid tới các post có cùng khu vực 

```python
last_visited_region = user_last_visited["region"][0]
data_same_region = df_scaled_data[df_scaled_data["region"] == last_visited_region].copy()
data_same_region
```

```python
from scipy.spatial.distance import cdist


data_same_region["score"] = cdist(data_same_region[cal_cols], user_last_visited[cal_cols], metric='euclidean')

df_result = data_same_region.sort_values("score", ascending=True)
recommended_id = df_result["id"].tolist()
print(recommended_id)
visited_id = user_last_visited["id"].tolist()[0]
recommended_id.remove(visited_id)
df_result
```

```python
recommended_post = pd.DataFrame(columns=df.columns)
for ordered_id in recommended_id:
    recommended_post = pd.concat([recommended_post, df[df["id"] == ordered_id]], ignore_index=True)
recommended_post
```

```python
user_last_visited[cal_cols] = scaler.inverse_transform(user_last_visited[cal_cols])
user_last_visited
```
